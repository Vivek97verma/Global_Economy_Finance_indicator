```python
# Strip any leading or trailing spaces from the column 
columns=[' Year ', ' IMF based exchange rate ', ' Population ',
       ' Currency ', '  Per capita GNI  ', '  Exports of goods and services  ',
       '  Final consumption expenditure  ',
       ' Gross National Income(GNI) in USD ',
       '  Gross Domestic Product (GDP)  ']
```

```python
df.columns = df.columns.str.strip()
print("Columns after stripping spaces:", df.columns.tolist())
```

```python
# fetch the maximum gdp of country w.r.t to financial year
df_max_gdp = df.groupby('Country')[['Gross Domestic Product (GDP)','Year']].max().reset_index()
df_max_gdp.sort_values(by='Gross Domestic Product (GDP)', ascending=False).head(10)
```

```python
# fetch the maximum GNI of country w.r.t to financial year
df_max_gdp = df.groupby('Country')[['Gross National Income(GNI) in USD','Year']].max().reset_index()
df_max_gdp.sort_values(by='Gross National Income(GNI) in USD', ascending=False).head(10)
```

```python
# per capita gross national income of each country
df_currency=df.groupby('Country')[['Currency','Per capita GNI']].max().reset_index()
df_currency.sort_values(by='Per capita GNI',ascending=False).head(10)
```

```python
# Countries with maximum population in Fy-2021
df_max_population=df.groupby('Country')[['Population','Year']].max().reset_index()
df_max_population.sort_values(by='Population',ascending=False).head(10)
```
